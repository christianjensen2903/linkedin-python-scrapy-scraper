```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the data companies.json
# and create a dataframe
df = pd.read_json("companies.json")
df
```

```python
# Replace not-found with NaN and -1 with NaN
df = df.replace("not-found", np.nan)
df = df.replace(-1, np.nan)

# Count the number of NaN values in each column
df.isna().sum()
```

```python
# Drop the rows with name as Confidential
df = df[df["name"] != "Confidential"]
```

```python
# Replace followers from the file companies_with_nan_followers.json

df2 = pd.read_json("companies_with_nan_followers.json")

for index, row in df2.iterrows():
    # Get the name of the company
    name = row["name"]
    # Get the followers of the company
    followers = row["followers"]
    # Find the row in df with the same name
    # and replace the followers
    df.loc[df["name"] == name, "followers"] = followers
```

```python
# Remove the rows where followers is NaN
df = df.dropna(subset=["followers"])
```

```python
# Count the number of each type of company
df["size"].value_counts()
```

```python
# Export the dataframe to a json file
df.to_json("companies_clean.json")
```

```python
df = pd.read_json("companies_clean.json")
```

```python
# Explode the posts column
df = df.explode("posts")
```

```python
# Convert the posts column into multiple columns
# The posts column is a dictionary
# We want to convert it into multiple columns
df = pd.concat([df.drop(["posts"], axis=1), df["posts"].apply(pd.Series)], axis=1)
```

```python
# Sort by likes
df = df.sort_values(by="comments", ascending=False)
```

```python
# replace not-found with NaN
df = df.replace("not-found", np.nan)
```

```python
# Drop columns where text, likes or comments is NaN
df = df.dropna(subset=["text", "likes", "comments"])
df
```

```python
# count the number of posts per content type
df["content_type"].value_counts()
```

```python
from langdetect import detect

def detect_language(text):
    try:
        return detect(text)
    except:
        return None
    
df['language'] = df['text'].apply(detect_language)

df = df[df['language'] == 'en']

df = df.drop(columns=['language'])
df
```

```python
# Export the dataframe to a json file
df.reset_index(inplace=True)
df.to_json("posts_clean.json", orient="records")
```
